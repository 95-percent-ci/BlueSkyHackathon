## Time Series Analysis 

**Important Consideration**

1. Stationarity
2. Trends
3. Autocorrelation

```python
tr_data["Rolling_avg"] = tr_data["T"].rolling('D').mean()

tr_data["Rolling_std"] = tr_data["T"].rolling('D').std()
```

```python
px.line(tr_data,y=['T', "Rolling_avg", "Rolling_std"])
```
